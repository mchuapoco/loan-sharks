```python
import pandas as pd
```

```python
data=pd.read_csv(".././LOANS_TRAIN.csv")
label=['loan_amnt','revol_bal','revol_util','pub_rec','pub_rec_bankruptcies','mort_acc','total_acc','open_acc','emp_length']
delete_label=['emp_title','title']
data.drop(delete_label,axis=1)
data_selected=data[label]
def remove_years(text):
    if(str(text)[:3]=="< 1"):
        return 0
    elif(str(text)[:2]=="10"):
        return 10
    else:
        return int(str(text)[:1]) if (str(text)[0]!="n") else 0
def remove_percent(text):
    return float(str(text)[:-1]) if str(text)!="nan" else text
data_selected['emp_length']=data_selected['emp_length'].apply(remove_years)
data_selected['revol_util']=data_selected['revol_util'].apply(remove_percent)
```
